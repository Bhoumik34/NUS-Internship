```python
import linktransformer as lt
import pandas as pd
import sklearn
```

```python
trdf= pd.read_csv('/content/ktrain.csv')
edf= pd.read_csv('/content/keval.csv')
tdf= pd.read_csv('/content/ktest.csv')
```

```python
train=trdf.drop([ 'Unnamed: 0','Neighbourhood', 'Village', 'Household'],axis=1)
eval= edf.drop([ 'Unnamed: 0','Neighbourhood', 'Village', 'Household'],axis=1)
test= tdf.drop([ 'Unnamed: 0','Neighbourhood', 'Village', 'Household'],axis=1)
```

```python
best_model_path,best_metric, label_map= lt.train_clf_model(train_data=train,val_data=eval,test_data=test,model="bert-base-multilingual-cased",on="Template",label_col_name="GP",data_dir=".",
                    lr=2e-5,batch_size=8,training_args={"num_train_epochs":10},model_save_dir="test_lt_clf", weighted_loss=True,weight_list=None,
                    wandb_log=True)
```

```python
df_clf_output = lt.classify_rows(test, on="Template", model=best_model_path)
```

```python
df_clf_output = df_clf_output.reset_index(drop=True)
df_clf_output
```

```python
df_clf_output['clf_preds_Template']=df_clf_output['clf_preds_Template'].astype(int)
```

```python
def calculate_false_positives_and_false_negatives(df_clf_output):
  fp=0 #false positive
  fn=0 #false negative
  for i in range(len(df_clf_output)):
      if df_clf_output['GP'][i]==0 and df_clf_output['clf_preds_Template'][i]==1:
         fp=fp+1
      if df_clf_output['GP'][i]==1 and df_clf_output['clf_preds_Template'][i]==0:
         fn=fn+1
  return fp,fn
```

```python
false_positives, false_negatives= calculate_false_positives_and_false_negatives(df_clf_output)
```

```python
print("false positives",false_positives)
print("false negatives", false_negatives)
```

```python
false_negatives = []
false_positives = []
for i in range(len(df_clf_output)):
    if df_clf_output['GP'][i]==0 and df_clf_output['clf_preds_Template'][i]==1:
       false_positives.append(df_clf_output['Template'][i])
    if df_clf_output['GP'][i]==1 and df_clf_output['clf_preds_Template'][i]==0:
       false_negatives.append(df_clf_output['Template'][i])

# Printing the text of false negatives
print("False Negatives:")
for text in false_negatives:
    print(text)

# Printing the text of false positives
print("\nFalse Positives:")
for text in false_positives:
    print(text)
```
