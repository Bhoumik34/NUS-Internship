```python
import linktransformer as lt
import pandas as pd
import sklearn
```

```python
import logging
logging.disable(logging.WARNING)
```

```python
train_df=pd.read_excel('/content/drive/MyDrive/train-k.xlsx')
eval_df=pd.read_excel('/content/drive/MyDrive/eval-k.xlsx')
test_df=pd.read_excel('/content/drive/MyDrive/test-k.xlsx')
```

```python
train_df= train_df.sample(frac=1)
eval_df= eval_df.sample(frac=1)
test_df= test_df.sample(frac=1)
```

```python
new_train_df=train_df.drop(['GP', 'Village', 'Household'],axis=1)
new_eval_df= eval_df.drop(['GP', 'Village', 'Household'],axis=1)
new_test_df= test_df.drop(['GP', 'Village', 'Household'],axis=1)
```

```python
best_model_path,best_metric, label_map= lt.train_clf_model(train_data=new_train_df,val_data=new_eval_df,test_data=new_test_df,model="bert-base-multilingual-cased",on="template",label_col_name="Neighbourhood",data_dir=".",
                    lr=2e-5,batch_size=16,training_args={"num_train_epochs":10},model_save_dir="test_lt_clf", weighted_loss=True,weight_list=None,
                    wandb_log=True)
```

```python
final_df= pd.read_csv('/content/merged_works_gsk.csv')
```

```python
#running inference on the original dataset
final_output = lt.classify_rows(final_df, on="work_name", model=best_model_path)
```

```python
final_output = final_output.reset_index(drop=True)
```
